```python
final_conv_list = []
final_dict = []
```

```python
for datapoint in range(total_dataPoints):
    final_conv_list = []
    for conversation_length in range(len(df['conversations'][datapoint]['MessageList'])):
        Msg_From = df['conversations'][datapoint]['MessageList'][conversation_length]['displayName']
        if Msg_From == None:
            final_conv_list.append(('Support',df['conversations'][datapoint]['MessageList'][conversation_length]['content']))
        else:
            final_conv_list.append(('User',df['conversations'][datapoint]['MessageList'][conversation_length]['content']))
    final_dict.append(final_conv_list)
    
```

```python
# final_df = pd.DataFrame(data=[[final_dict.keys()],[final_dict.items()]], columns = ['Index', 'Message'])
final_df = pd.DataFrame([final_dict]).T
```

```python
final_df.to_csv("conversations.csv")
```
